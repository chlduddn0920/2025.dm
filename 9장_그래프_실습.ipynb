```python
# 쉬운 DFS (깊이 우선 탐색)
def dfs(graph, start):
    visited = []       # 방문한 노드들을 저장
    stack = [start]    # 다음에 방문할 노드를 담는 스택

    while stack:
        node = stack.pop()
        if node not in visited:
            print(node, end=' ')
            visited.append(node)
            # 인접 노드를 거꾸로 넣으면 사전순 탐색됨
            for neighbor in reversed(graph[node]):
                if neighbor not in visited:
                    stack.append(neighbor)
    return visited

# 예시 그래프
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

print("DFS 탐색 순서:")
dfs(graph, 'A')
```

```python
from collections import deque

# 예시 그래프 (사람들 관계)
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# -------------------------------
# DFS : 깊이 우선 탐색
# -------------------------------
def dfs(graph, start):
    visited = []
    stack = [start]
    print("DFS 탐색 순서:")

    while stack:
        node = stack.pop()
        if node not in visited:
            print(node, end=' ')
            visited.append(node)
            for neighbor in reversed(graph[node]):
                if neighbor not in visited:
                    stack.append(neighbor)
    print("\n")
    return visited

# -------------------------------
# BFS : 너비 우선 탐색
# -------------------------------
def bfs(graph, start):
    visited = []
    queue = deque([start])
    print("BFS 탐색 순서:")

    while queue:
        node = queue.popleft()
        if node not in visited:
            print(node, end=' ')
            visited.append(node)
            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append(neighbor)
    print("\n")
    return visited

# -------------------------------
# 감염 시뮬레이션 (BFS 2단계)
# -------------------------------
def infection_simulation(graph, start, max_step=2):
    queue = deque([(start, 0)])
    infected = {start}
    print(f"감염 시작: {start}")
    print(f"{max_step}단계까지 전염 경로:")

    while queue:
        node, step = queue.popleft()
        if step < max_step:
            for friend in graph[node]:
                if friend not in infected:
                    infected.add(friend)
                    print(f"{' ' * (step*2)}{node} → {friend} (단계 {step+1})")
                    queue.append((friend, step + 1))
    print("\n전체 감염자:", infected)
    return infected

# 실행 부분
dfs(graph, 'A')
bfs(graph, 'A')
infection_simulation(graph, 'A', max_step=2)
```

```python
# ========================= 과제 메타 =========================
# 프롬프트(요약): "인접 리스트 그래프에서 DFS/BFS 탐색 순서와
#                 감염 시뮬레이션(거리 k 이내, 본인 제외)을 소/중/대 3케이스로 구현"
# 사용 AI 서비스: GPT-5 (ChatGPT)
# ============================================================

from collections import deque, defaultdict
import random

# ------------------------------------------------------------
# 공통 유틸
# ------------------------------------------------------------
def dfs(graph, start):
    """반복형 DFS (스택). 방문 순서 리스트 반환."""
    order, visited, stack = [], set(), [start]
    while stack:
        node = stack.pop()
        if node in visited:
            continue
        visited.add(node)
        order.append(node)
        # LIFO 특성상 역순으로 넣으면 사람이 보기 좋은 순서
        for nb in reversed(graph.get(node, [])):
            if nb not in visited:
                stack.append(nb)
    return order

def bfs(graph, start):
    """반복형 BFS (큐). 방문 순서 리스트 반환."""
    order, visited, q = [], set([start]), deque([start])
    while q:
        cur = q.popleft()
        order.append(cur)
        for nb in graph.get(cur, []):
            if nb not in visited:
                visited.add(nb)
                q.append(nb)
    return order

def infection_within_k(graph, patient_zero, k=2):
    """
    BFS로 거리<=k 감염자 레벨별 수집. 본인은 결과에서 제외.
    return: {level: [nodes...]} (level=1..k, 정렬)
    """
    visited = {patient_zero: 0}
    q = deque([patient_zero])
    by_level = defaultdict(list)

    while q:
        cur = q.popleft()
        d = visited[cur]
        if d == k:
            continue
        for nb in graph.get(cur, []):
            if nb not in visited:
                visited[nb] = d + 1
                q.append(nb)
                if visited[nb] >= 1:
                    by_level[visited[nb]].append(nb)

    # 보기 좋게 정렬
    for lvl in by_level:
        by_level[lvl].sort()
    return dict(sorted(by_level.items()))

def print_case(title, graph, start, k):
    print(f"\n================ {title} ================")
    print("[그래프(인접 리스트)]")
    for u in sorted(graph):
        print(f"  {u}: {graph[u]}")
    print("\n[DFS 순서]")
    print("  " + " -> ".join(dfs(graph, start)))
    print("[BFS 순서]")
    print("  " + " -> ".join(bfs(graph, start)))

    by_level = infection_within_k(graph, start, k=k)
    print(f"\n[감염 시뮬레이션] 환자0={start}, 최대 {k}단계")
    if not by_level:
        print("  (감염 확산 없음)")
    else:
        total = 0
        for lvl in sorted(by_level):
            nodes = by_level[lvl]
            total += len(nodes)
            print(f"  - {lvl}단계({len(nodes)}명): " + ", ".join(nodes))
        print(f"  ▶ 총 감염자 수(본인 제외): {total}")

# ------------------------------------------------------------
# [소] 노드 10 내외, k=2 — 우리 조 이름 포함
# ------------------------------------------------------------
graph_small = {
    '이연우': ['이윤호', '최영우', '박지윤'],
    '이윤호': ['이연우', '구다은', '정민'],
    '최영우': ['이연우', '구다은', '김하늘'],
    '구다은': ['이윤호', '최영우', '전우성'],
    '박지윤': ['이연우', '전우성'],
    '전우성': ['박지윤', '구다은'],
    '정민': ['이윤호'],
    '김하늘': ['최영우'],
}
start_small = '이연우'
print_case("소형 케이스 (노드≈8, k=2)", graph_small, start_small, k=2)

# ------------------------------------------------------------
# [중] 노드 100 내외, k=10
#   - 연결성 보장 위해 일단 선형 체인 + 임의의 보조 간선 추가
# ------------------------------------------------------------
def build_medium_graph(n=100, seed=7):
    random.seed(seed)
    adj = {f"N{i}": [] for i in range(n)}
    # 선형 연결
    for i in range(n-1):
        a, b = f"N{i}", f"N{i+1}"
        adj[a].append(b); adj[b].append(a)
    # 추가 간선
    for _ in range(n*2):
        u, v = random.randrange(n), random.randrange(n)
        if u == v:
            continue
        a, b = f"N{min(u,v)}", f"N{max(u,v)}"
        if b not in adj[a]:
            adj[a].append(b); adj[b].append(a)
    # 정렬
    for k in adj:
        adj[k] = sorted(adj[k])
    return adj

graph_med = build_medium_graph(100, seed=11)
start_med = "N0"
print_case("중형 케이스 (노드≈100, k=10)", graph_med, start_med, k=10)

# ------------------------------------------------------------
# [대] 노드 1000 내외, k=50 — 격자 그래프
# ------------------------------------------------------------
def build_grid_graph(rows=25, cols=40):
    adj = {}
    for r in range(rows):
        for c in range(cols):
            u = f"G_{r}_{c}"
            neigh = []
            if r+1 < rows: neigh.append(f"G_{r+1}_{c}")
            if c+1 < cols: neigh.append(f"G_{r}_{c+1}")
            if r-1 >= 0:   neigh.append(f"G_{r-1}_{c}")
            if c-1 >= 0:   neigh.append(f"G_{r}_{c-1}")
            adj[u] = neigh
    return adj

graph_big = build_grid_graph(25, 40)  # 25*40=1000
start_big = "G_0_0"
# 출력이 너무 길어지니까 탐색 순서는 앞부분만 슬라이스해서 보자
print(f"\n================ 대형 케이스 (노드≈1000, k=50) ================\n"
      f"[그래프 요약] 시작노드={start_big}, 이웃={graph_big[start_big][:4]} ...")
dfs_big = dfs(graph_big, start_big)[:30]
bfs_big = bfs(graph_big, start_big)[:30]
print("[DFS 순서(앞 30개)]")
print("  " + " -> ".join(dfs_big) + " -> ...")
print("[BFS 순서(앞 30개)]")
print("  " + " -> ".join(bfs_big) + " -> ...")

by_level_big = infection_within_k(graph_big, start_big, k=50)
# 1~5단계만 요약
print("\n[감염 시뮬레이션] 환자0=G_0_0, 최대 50단계 (요약 1~5단계)")
total = 0
for lvl in range(1, 6):
    nodes = by_level_big.get(lvl, [])
    total += len(nodes)
    preview = ", ".join(nodes[:10])
    if nodes:
        print(f"  - {lvl}단계({len(nodes)}명): {preview}{' ...' if len(nodes)>10 else ''}")
    else:
        print(f"  - {lvl}단계(0명): (없음)")
print(f"  ▶ (요약) 1~5단계 합계: {total}")
```
